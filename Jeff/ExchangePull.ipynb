```python
import numpy as np
import pandas as pd
import requests
import time
import winsound
```

```python
path = "countrycode.csv"

exdtdf = pd.read_csv(path, encoding = "ISO-8859-1").dropna()
```

```python
countrylst = list(exdtdf['ALL'])
```

```python
url = "https://www.alphavantage.co/query?function=CURRENCY_EXCHANGE_RATE"
apikey="V8UXD8118Z1PW4MZ"
datatype ="json"
lst2=[]
cto = 'USD'
```

```python
for cfrom in countrylst:
    time.sleep(10) 
    cfrom = cfrom
    try:
        rs = requests.get(f"{url}&from_currency={cfrom}&to_currency={cto}&apikey={apikey}&datatype={datatype}").json()
        lst =[]
        lst.append (rs['Realtime Currency Exchange Rate']['1. From_Currency Code'])
        lst.append (rs['Realtime Currency Exchange Rate']['2. From_Currency Name'])
        lst.append (rs['Realtime Currency Exchange Rate']['3. To_Currency Code'])
        lst.append (rs['Realtime Currency Exchange Rate']['4. To_Currency Name']) 
        lst.append (rs['Realtime Currency Exchange Rate']['5. Exchange Rate']) 
        lst.append (rs['Realtime Currency Exchange Rate']['6. Last Refreshed'])  
        lst2.append(lst)
        print({cfrom})
    except:
        print(f'----->{cfrom}')
        continue
duration = 1000  # millisecond
freq = 440  # Hz
winsound.Beep(freq, duration)        
```

```python
df = pd.DataFrame(lst2, columns=['fromcode','fromname','tocode','toname','rate','lastrefreshed'])

```
