```python
import numpy as np
import pandas as pd
import requests
import time
```

```python
path = "Fortune_500.csv"

tickerdf = pd.read_csv(path, encoding = "ISO-8859-1").dropna()
symbollst = list(tickerdf['Ticker Symbol'])
```

```python
symbollst
```

```python
lst2 = []
url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY"
apikey="V8UXD8118Z1PW4MZ"
datatype ="json"
```

```python
for sym in symbollst:
    time.sleep(10)    
    symbol = sym
    try:
        rs = requests.get(f"{url}&symbol={sym}&apikey={apikey}&datatype={datatype}").json()
        dts =rs['Time Series (Daily)'].keys()
        for dt in dts:
            lst =[]
            lst.append (symbol)
            lst.append (dt)
            lst.append (rs['Time Series (Daily)'][dt]['1. open'])
            lst.append (rs['Time Series (Daily)'][dt]['2. high'])
            lst.append (rs['Time Series (Daily)'][dt]['3. low'])
            lst.append (rs['Time Series (Daily)'][dt]['4. close']) 
            lst.append (rs['Time Series (Daily)'][dt]['5. volume'])  
            lst2.append(lst)
    except:
        continue
df = pd.DataFrame(lst2, columns=['symbol','date','open','high','low','close','volume'])
```
