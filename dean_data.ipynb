```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import math
import pprint
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)


review_company = 'Apple'
url = "https://www.indeed.com/cmp/" + review_company + "/reviews"
browser.visit(url)
html = browser.html
#html
```

```python
soup = BeautifulSoup(html, "html.parser")

pages_text = soup.find('span', {'class':'cmp-filter-result'}).text.strip().split()
pages_text
page_count = math.ceil(int(pages_text[1].replace(",", ""))/21)
```

```python
column_list = ["company", "rating_number", "reviewer", "review_job_title", "review_job_location", \
                     "review_text", "review_date_created", "page_retrieved"]
export_df = pd.DataFrame(columns=column_list)

for i in range(1, page_count):

    for comment in soup.find_all('div', {'class':'cmp-review-container'}):
        rating_number = comment.find('div', {'class':'cmp-ratingNumber'}).text
        review_title = comment.find('div', {'class':'cmp-review-title'}).text
        reviewer = comment.find('span', {'class':'cmp-reviewer'}).text
        review_job_title = comment.find('span', {'class':'cmp-reviewer-job-title'}).text
        review_job_location = comment.find('span', {'class':'cmp-reviewer-job-location'}).text
        review_text = comment.find('span', {'class':'cmp-review-text'}).text
        review_date_created = comment.find('span', {'class':'cmp-review-date-created'}).text
        
        export_df = export_df.append(pd.DataFrame([[review_company, rating_number, reviewer, review_job_title, review_job_location,
                     review_text, review_date_created, i]], columns=column_list))
        print(rating_number)
        print(review_title)
        print(reviewer)
        print(review_job_title)
        print(review_job_location)
        print(review_text[1:60])
        print(review_date_created)
        print('Page ' + str(i))
    
    browser.click_link_by_partial_text('Next')
```

```python
export_file_name = './data/' + review_company + '_reviews.csv'

#export_df = export_df.set_index('company')
export_df.to_csv(export_file_name)
```
